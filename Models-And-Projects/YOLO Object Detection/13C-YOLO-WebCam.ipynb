```python
# Step 1: Import the necessary library
from ultralytics import YOLO

# Step 2: Load a pretrained YOLOv8 pose model
# You can use 'yolov8n-pose.pt' (nano) or other versions like 'yolov8s-pose.pt'
model = YOLO('yolov8n-pose.pt')

# Step 3: Run the model on your webcam
# Passing 0 tells it to use the default webcam. You can also use 1, 2, etc. for other cameras.
model.predict(source=0, show=True)
```
