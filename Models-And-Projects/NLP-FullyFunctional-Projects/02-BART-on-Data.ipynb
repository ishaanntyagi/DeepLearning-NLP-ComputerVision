```python
import pickle 
from transformers import pipeline
```

```python
with open("preprocessed_contracts.pkl", "rb") as file:
    preprocessed_contracts = pickle.load(file)
```

```python
if isinstance(preprocessed_contracts[0], list):
    texts = [" ".join(words) for words in preprocessed_contracts]
else:
    texts = preprocessed_contracts
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
for i in range(len(texts)):
    input_text = texts[i][:1024] 
    summary = summarizer(input_text, max_length=100, min_length=30, do_sample=False)
    print(summary[0]['summary_text'])
    print("-------------------------------------------------------------------------")
```
